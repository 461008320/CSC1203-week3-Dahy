```python
score = float(input("Enter student score: "))


if score >= 60:
    print("Status: Passed")
else:
    print("Status: Failed")


if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"


match grade:
    case "A":
        message = "Excellent work!"
    case "B":
        message = "Well done!"
    case "C":
        message = "Good job."
    case "D":
        message = "You passed, but need more practice."
    case "F":
        message = "Better luck next time."

print(f"Grade: {grade}")
print(f"Message: {message}")
```
